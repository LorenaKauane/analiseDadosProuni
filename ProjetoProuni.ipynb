<h1>Prouni</h1>

<p>O Programa Universidade para Todos - Prouni tem como finalidade a concessão de bolsas de estudo integrais e parciais em cursos de graduação e sequenciais de formação específica, em instituições de ensino superior privadas. </p>
<p>Criado pelo Governo Federal em 2004 e institucionalizado pela Lei nº 11.096, em 13 de janeiro de 2005 oferece, em contrapartida, isenção de tributos àquelas instituições que aderem ao Programa. Dirigido aos estudantes egressos do ensino médio da rede pública ou da rede particular na condição de bolsistas integrais, com renda familiar per capita máxima de três salários mínimos, o Prouni conta com um sistema de seleção informatizado e impessoal, que confere transparência e segurança ao processo. </p>
<p>Os candidatos são selecionados pelas notas obtidas no Exame Nacional do Ensino Médio - Enem conjugando-se, desse modo, inclusão à qualidade e mérito dos estudantes com melhores desempenhos acadêmicos.</p>

<h1>Import de bibliotecas</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

<b>Leitura do arquivo e informações dos dados e colunas</b>

```python
df = pd.read_csv('cursos-prouni-original.csv', encoding = "utf-8")
```

<h1>Limpando colunas nulas</h1>

```python
#Clean
df['mensalidade'].fillna(0, inplace=True)
df['bolsa_integral_cotas'].fillna(0, inplace=True)
df['bolsa_integral_ampla'].fillna(0, inplace=True)
df['bolsa_parcial_cotas'].fillna(0, inplace=True)
df['bolsa_parcial_ampla'].fillna(0, inplace=True)
df['nota_integral_ampla'].fillna(0, inplace=True)
df['nota_integral_cotas'].fillna(0, inplace=True)
df['nota_parcial_ampla'].fillna(0, inplace=True)
df['nota_parcial_cotas'].fillna(0, inplace=True)
```

```python
df.head(5)
```

<h1>Perguntas básicas</h1>

<b>Quais são os estados que mais tem universidades no prouni?

```python
df['uf_busca'].value_counts().sort_values(ascending=True).plot(kind = 'barh',figsize=(8,10))
```

<b>Quais são os 50 cursos que mais aparecem no dataset?

```python
plt.figure(figsize=(8,10))

descending_order = df['nome'].value_counts().head(50).sort_values(ascending=False).index

sns.countplot(
    y=df.nome, 
    order=descending_order, 
    orient="v")

plt.yticks(fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

<b>Qual e a proporção da quantidade de cursos por estados onde o curso é administração ?

```python
totalUfPorAdm = df[df["nome"]== "Administração"].groupby("uf_busca")["cidade_busca"] \
    .count() \
    .reset_index(name='cidade_busca') \
    .sort_values(["cidade_busca"],ascending=False)
totalUfPorAdm
```

```python
totalUf = df['uf_busca'].value_counts()

f, ax = plt.subplots(figsize=(6, 15))

sns.set_color_codes("pastel")
sns.barplot(x=totalUf.values, y=totalUf.index, data=df,
             label="UF", color="b")
sns.set_color_codes("muted")

sns.barplot(x="cidade_busca", y="uf_busca", data=totalUfPorAdm,
            label="Cursos ADM", color="b")

ax.legend(ncol=2, frameon=True)
ax.set(xlim=(0, 800), ylabel="",
       xlabel="Proporção do curso de Administração nos estados")
sns.despine(left=True, bottom=True)
```

<b> Mensalidade > ou < nos cursos de ADM

```python
def sortAdmMensalidade(bool, total):
    dfAdm =  df[(df["nome"]== "Administração")].sort_values(by='mensalidade', ascending=bool)
    return dfAdm.head(total)
```

```python
aux = sortAdmMensalidade(True,23)

plt.figure(figsize=(8,10))

plt.title("20 Cidades onde a mensalidade de Administração e menor")
plt.plot(aux["mensalidade"], aux["cidade_busca"],'r--')

plt.yticks(fontsize=12)
plt.xticks(fontsize=10)

plt.xlabel("mensalidade")
plt.xlim(165, 200)
```

```python
aux = sortAdmMensalidade(False,48)

plt.figure(figsize=(8,10))

plt.title("20 Cidades onde a mensalidade de Administração e maior")
plt.plot(aux["mensalidade"], aux["cidade_busca"],'r--')

plt.yticks(fontsize=12)
plt.xticks(fontsize=10)

plt.xlabel("mensalidade")
plt.xlim(1000, 7000)
```

<b>Grau e as medias da mensalidade
